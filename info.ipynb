What does info() do?
The function info() gives you a quick summary of your whole dataset (DataFrame).

Think of it like asking:

‚ÄúHey Pandas, can you tell me what's in this table?‚Äù

‚úÖ Yes! info() tells you:

How many rows and columns you have

The names of the columns

The data types of each column (like int, float, object, etc.)

How many non-null (not empty) values are in each column

How much memory is used



Let‚Äôs make a small table (DataFrame) of student data:

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [20, 21, 19, 22, None],
    'Score': [85, 90, 95, 88, 92]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.info()
```

| üî¢ Part                    | üß∑ Meaning                                                         |
| -------------------------- | ------------------------------------------------------------------ |
| `RangeIndex: 5 entries`    | There are 5 rows, from 0 to 4                                      |
| `Data columns: 3`          | There are 3 columns: Name, Age, Score                              |
| `Name ‚Äì 5 non-null object` | Column `Name` has 5 filled values, and the type is text (`object`) |
| `Age ‚Äì 4 non-null float64` | Column `Age` has **1 missing** value (since only 4 are non-null)   |
| `Score ‚Äì 5 non-null int64` | Column `Score` is full and contains integers                       |
| `Memory usage`             | How much memory your DataFrame takes up (important for big data)   |
