```python
import pandas as pd

# 创建一个简单的 DataFrame
data = {
    '姓名': ['张三', '李四', '王五', '赵六', '钱七', '孙八', '周九'],
    '年龄': [25, 30, 35, 40, 45, 50, 55],
    '城市': ['北京', '上海', '广州', '深圳', '杭州', '成都', '武汉']
}

df = pd.DataFrame(data)

# 使用 head() 函数查看 DataFrame 的前 5 行
print("默认查看前 5 行：")
print(df.head())

# 使用 head() 函数查看 DataFrame 的前 3 行
print("\n查看前 3 行：")
print(df.head(3))

# 如果想查看更多行，比如前 10 行，可以这样做
# 注意：如果 DataFrame 的行数少于请求的行数，它会显示所有行
print("\n尝试查看前 10 行：")
print(df.head(10))
```
